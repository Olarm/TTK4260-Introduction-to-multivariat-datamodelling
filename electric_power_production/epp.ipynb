```python
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
x_data = pd.read_csv('x.txt', sep=";")
y_data = pd.read_csv('y.txt', sep=";")
x1 = x_data[0:13150]
y1 = y_data[0:13150]

y1 = y1.drop(["Row"], axis=1)
data = pd.concat([x1, y1], axis=1, sort=False)
```

```python
data = pd.get_dummies(data, columns=["IsDayBin"])
data = data.drop(["IsDayBin_Night"], axis=1)
```

```python
data.head()
```

```python
ax = sns.lineplot(x="Month", y="Temperature",  hue="IsDayBin_Day", data=data)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(10, 10))
sns.despine(f, left=True, bottom=True)
sns.scatterplot(x="WindSpeed", y="WindPower",
                palette="coolwarm",
                hue="IsDayBin_Day",
                linewidth=0,
                data=data, ax=ax)

plt.show()
```
